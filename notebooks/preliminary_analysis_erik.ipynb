```python
import os
import pandas as pd

def get_parquet_files(root_folder):
    parquet_files = []
    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.endswith(".parquet"):
                parquet_files.append(os.path.join(root, file))
    return parquet_files

root_folder = '../test/'
parquet_files = get_parquet_files(root_folder)
print(parquet_files)
transactions = pd.concat([pd.read_parquet(parquet_file) for parquet_file in parquet_files])
transactions.reset_index(inplace=True)
transactions.head()
```

```python
import pandas as pd

# Ensure 'dollar_value' column is numeric
transactions['dollar_value'] = pd.to_numeric(transactions['dollar_value'], errors='coerce')

# Drop rows with NaN values in 'dollar_value' column
transactions = transactions.dropna(subset=['dollar_value'])

# Show the highest value of the 'dollar_value' column
max_dollar_value = transactions['dollar_value'].max()
print(f"Highest dollar value: {max_dollar_value}")

# Show the lowest value of the 'dollar_value' column
min_dollar_value = transactions['dollar_value'].min()
print(f"Lowest dollar value: {min_dollar_value}")

# Get the index of the row with the highest dollar_value
max_dollar_value_index = transactions[transactions['dollar_value'] == max_dollar_value].index

# Get the row with the highest dollar_value
max_dollar_value_row = transactions.loc[max_dollar_value_index]

# Get the index of the row with the lowest dollar_value
min_dollar_value_index = transactions[transactions['dollar_value'] == min_dollar_value].index

# Get the row with the lowest dollar_value
min_dollar_value_row = transactions.loc[min_dollar_value_index]


# Print the rows
print("Row(s) with the highest dollar value:")
print(max_dollar_value_row)

print("\nRow(s) with the lowest dollar value:")
print(min_dollar_value_row)
```
